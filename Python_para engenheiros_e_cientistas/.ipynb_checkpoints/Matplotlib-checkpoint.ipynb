## 10.1.1. Pyplot e elementos gráficos

### Agenda

- Uso no Jupyter notebook: **%matplotlib notebook**

- Usando o pyplot: **import matplotlib.pyplot as plt**


- **Marcadores** : https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.plot.html
- Elementos gráficos: Marcadores, títulos, eixos, grade e legenda

```python
import matplotlib.pyplot as plt

# Comando necessário para ambientes que não tenha suporte nativo aos elementos(prompts)
%matplotlib notebook
```

```python
import numpy as np

x_array = np.linspace(0,2*np.pi,500)
y_array = np.sin(x_array)

plt.plot(x_array, y_array,'r')
```

```python
xb_array = np.linspace(0,2*np.pi,500)
yb_array = np.sin(xb_array + np.pi/2)

plt.plot(xb_array, yb_array,'k')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
## Prática 2

fig, ax = plt.subplots()

ax.plot(x_array, y_array, 'k',label='fun a')
ax.plot(xb_array, yb_array, 'r', label='fun b')
ax.grid()
plt.legend()

ax.set_title('Função seno')
ax.set(title='Função seno', xlabel='x',ylabel='y')
```

## 10.1.2. Subplots

- Dica: subplots são armazenados em arrays numpy
- https://matplotlib.org/gallery/subplots_axes_and_figures/subplot.html

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
import numpy as np

xa_array = np.linspace(0,2,500)
ya_array = np.exp(xa_array)
```

```python
fig, (ax1, ax2) = plt.subplots(2,2)

ax1[0].plot(xa_array, ya_array,'r')
ax1[1].plot(xa_array, ya_array,'r')
ax2[0].plot(xa_array, ya_array,'r')
ax2[1].plot(xa_array, ya_array,'r')

ax1[0].grid()
plt.show()
```

```python
plt.figure()

plt.subplot(221)
plt.plot(xa_array, ya_array)
plt.grid()

plt.subplot(224)
plt.plot(xa_array, ya_array)
plt.grid()
```

## 10.1.3. Escala Log

https://matplotlib.org/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py

```python
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np

x_array = np.linspace(0,10,500)
y_array = np.exp(x_array)

plt.figure()
plt.suptitle('Comparação escalas')

plt.subplot(211)
plt.plot(x_array, y_array, 'r')
plt.yscale('log')
plt.ylabel('Y log')
plt.grid()

plt.subplot(212)
plt.plot(x_array, y_array, 'k')
plt.ylabel('Y linear')
plt.grid()

plt.show()
```
