### Agenda

- Uso no Jupyter notebook: **%matplotlib notebook**

- **Marcadores** : https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.plot.html

````python
import matplotlib.pyplot as plt


# Comando necessário para ambientes que não tenha suporte nativo aos elementos(prompts)
%matplotlib notebook
````

````python
import numpy as np

x_array = np.linspace(0, 2 * np.pi, 500)
y_array = np.sin(x_array)

plt.plot(x_array, y_array, "r")
````

````python
xb_array = np.linspace(0, 2 * np.pi, 500)
yb_array = np.sin(xb_array + np.pi / 2)

plt.plot(xb_array, yb_array, "k")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
````

````python
## Prática 2

fig, ax = plt.subplots()

ax.plot(x_array, y_array, "k", label="fun a")
ax.plot(xb_array, yb_array, "r", label="fun b")
ax.grid()
plt.legend()

ax.set_title("Função seno")
ax.set(title="Função seno", xlabel="x", ylabel="y")
````

## 10.1.3. Escala Log

https://matplotlib.org/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x_array = np.linspace(0,10,500)
y_array = np.exp(x_array)

plt.figure()
plt.suptitle('Comparação escalas')

plt.subplot(211)
plt.plot(x_array, y_array, 'r')
plt.yscale('log')
plt.ylabel('Y log')
plt.grid()

plt.subplot(212)
plt.plot(x_array, y_array, 'k')
plt.ylabel('Y linear')
plt.grid()

plt.show()
````

## 10.1.5. Barras

https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html

matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
nomes = ["A", "B", "C", "D"]
valores = [20, 50, 10, 30]
valores2 = [10, 20, 30, 20]

plt.bar(nomes, valores, color="c")
plt.bar(nomes, valores2, color="r")
plt.show()
````

````python
fig, ax = plt.subplots()

x = np.arange(4)
large = 0.4

rect1 = ax.bar(x + large / 2, valores, width=large, label="gráfico 1")
rect2 = ax.bar(x - large / 2, valores2, width=large, label="gráfico 2")


ax.set_xticks(x)
ax.set_xticklabels(nomes)

plt.legend()
plt.show()
````

## 10.1.7 - Gráfico de superfície
**Estrutura**

```python
# Criar a figura 3d
>>>import matplotlib.pyplot as plt
>>>from matplotlib import cm #importa a ferramenta de colormap
>>>figura = plt.figure().gca(projection = '3d') #gca é um método para setar o eixo de coordenadas da figura
>>>figura.plot_surface(X,Y,Z) #método para fazer o plot de superfície

# O numpy fornece uma função para criar um mapa de coordenadas
>>> import numpy as np
>>> np.meshgrid(arr1,arr2) #cria um mapa de coordenadas    
```

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
````

````python
A = np.arange(3)
malha = np.meshgrid(A, A)  # plano com coordenadas: (a, b)

z = np.array([[0, 0, 0], [0, 0, 0], [1, 2, 3]])
````

````python
fig = plt.figure()
ax = fig.gca(projection="3d")

surf = ax.plot_surface(malha[0], malha[1], z)
````

````python
pi = np.linspace(0, np.pi, 50)
malha_2 = np.meshgrid(pi, pi)

z_2 = np.sin(malha_2[0]) * np.sin(malha_2[1])
````

````python
fig = plt.figure()
ax = fig.gca(projection="3d")

surf = ax.plot_surface(malha_2[0], malha_2[1], z_2, cmap=cm.coolwarm)
````
